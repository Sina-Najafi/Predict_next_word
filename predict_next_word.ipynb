```python
# Read the input text file
import tensorflow as tf
with open('corpus.txt') as f:
    corpus = f.readlines()
print(corpus[:3])
```

##Data preprocessing

```python
# let's get rid of \n
data =[]
helper_str = ''
for sentence in corpus:
  for character in sentence:
    if (character == '\n'):
      character = ''
    helper_str = helper_str + character
  data.append(helper_str)
  helper_str = ''
print(data[:3])
print(len(data))

# to get the number of unique tokens, I change the whole data to a single string
text = ' '.join(data)
print(len(text))
```

```python
print(data[:100])
```

```python
tokens = text.split()
print(tokens[:50])
print(list(set(tokens))[:50])
print(len(tokens))
print(len(set(tokens))) # number of unique tokens
```

```python
length = 2 + 1 # based on the last 2 words it predicts the 3rd word
lines = []
for i in range(length, len(tokens)):
  seq = tokens[i-length:i]
  line = ' '.join(seq)
  lines.append(line)
  
print(len(lines))
```

```python
print(lines[0])
print(lines[1])
print(lines[2])
print(lines[:3])
```

```python
lines_test = []
for i in range(length, len(tokens_test)):
  seq_test = tokens_test[i-length:i]
  line_test = ' '.join(seq_test)
  lines_test.append(line_test)
  
print(len(lines_test))
```

```python
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
import pandas as pd

tokenizer = Tokenizer(3600, oov_token = 'oov')
tokenizer.fit_on_texts(data)
sequences = tokenizer.texts_to_sequences(lines)
sequences_test = tokenizer.texts_to_sequences(lines_test)
print([sequences[0]])

# I'll just get rid of the ones with shape problems!
new_seq = []
for item in sequences:
  if len(item) == length:
    new_seq.append(item)

new_seq_test = []
for item in sequences_test:
  if len(item) == length:
    new_seq_test.append(item)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding
model = Sequential()
model.add(Embedding(vocab_size, 50, input_length = length - 1))
model.add(LSTM(20, return_sequences = True))
model.add(LSTM(20))
model.add(Dense(20, activation = 'relu'))
model.add(Dense(vocab_size, activation = 'softmax'))
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy',optimizer = tf.keras.optimizers.Adam(learning_rate=0.01), metrics = ['accuracy'])
model.fit(X, y, batch_size = 32, epochs = 50, validation_data=(X_test,y_test))
```

```python
# predicting just one possible word
def predict_next_word(model, tokenizer, text):
  sequence = tokenizer.texts_to_sequences([text])
  sequence = np.array(sequence)

  y_pred = np.argmax(model.predict(sequence))
  for key, value in tokenizer.word_index.items():
    if value == y_pred:
      predicted_word = key
      break
  print(predicted_word)
  return predicted_word
  
```

```python
# predicting 3 possible words
def predict_next_words(model, tokenizer, text):
  sequence = tokenizer.texts_to_sequences([text])
  sequence = np.array(sequence)
  probs = model.predict(sequence)
  probs = probs.tolist()
  n = 3
  result = [probs[0].index(i) for i in sorted(probs[0], reverse=True)][:n]
  first = result[0] 
  second = result[1] 
  third = result[2]
  result = [first, second , third]
  counter = 0
  predicted_words = []

  for item in result:
    for key, value in tokenizer.word_index.items():
      if value == item:
        predicted_words.append(key)
        counter += 1
      if counter == 3:       
        break       
  print(predicted_words)
  return predicted_words
  
```
